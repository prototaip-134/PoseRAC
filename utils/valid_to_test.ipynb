```python
import pandas as pd
```

```python
df = pd.read_csv('./valid.csv')
```

```python
df.head()
```

```python
# Function to calculate count
def calculate_count(row):
    l_columns = [col for col in df.columns if col.startswith('L')]
    non_nan_count = sum(pd.notna(row[col]) for col in l_columns)
    return non_nan_count // 2

# Apply function to create new column
df['count'] = df.apply(calculate_count, axis=1)

# Insert count column after 'name'
df = df[['Unnamed: 0', 'type', 'name', 'count'] + [col for col in df.columns if col not in ['Unnamed: 0', 'type', 'name', 'count']]]
```
